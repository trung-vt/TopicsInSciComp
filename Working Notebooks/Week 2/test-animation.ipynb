```python

# MOST IMPORTANT THING IN THE WORLD!!!
# https://decovar.dev/blog/2020/09/05/matplotlib-animation-macos/
# %matplotlib Tk
# %matplotlib Notebook
# %matplotlib notebook

# help(plt.plot)


# https://matplotlib.org/stable/users/explain/animations/animations.html

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.animation as animation

print("backend", plt.rcParams["backend"])
# #plt.rcParams["backend"] = "TkAgg" # doesn't actually set the backend
matplotlib.use("TkAgg")
# matplotlib.use("notebook") # ValueError: 'notebook' is not a valid value for backend; supported values are ['GTK3Agg', 'GTK3Cairo', 'GTK4Agg', 'GTK4Cairo', 'MacOSX', 'nbAgg', 'QtAgg', 'QtCairo', 'Qt5Agg', 'Qt5Cairo', 'TkAgg', 'TkCairo', 'WebAgg', 'WX', 'WXAgg', 'WXCairo', 'agg', 'cairo', 'pdf', 'pgf', 'ps', 'svg', 'template']
# matplotlib.use("pdf") # UserWarning: Matplotlib is currently using pdf, which is a non-GUI backend, so cannot show the figure.
# matplotlib.use("QtAgg")
# matplotlib.use("module://ipympl")
# matplotlib.use("nbAgg")
# matplotlib.use("WebAgg")

print("backend", plt.rcParams["backend"])

# fig, ax = plt.subplots()
# t = np.linspace(0, 3, 40)
# g = -9.81
# v0 = 12
# z = g * t**2 / 2 + v0 * t

# v02 = 5
# z2 = g * t**2 / 2 + v02 * t

asteroid_locations= np.loadtxt("data/asteroid_locations.dat")
# for location in asteroid_locations :
#     circle=plt.Circle(location, 0.5 , color='r')
#     ax.add_artist(circle)
# plt.pause(0.3)
print(asteroid_locations)

fig = plt.figure()
ax = plt.gcf().gca()
plt.axis([0,10.5,-4,4])
major_ticks = np.arange(0, 11, 1)
ax.set_xticks(major_ticks)
plt.xlabel("time / x position")
plt.ylabel("y position")
ax.set_aspect('equal')
ax.grid(which='major', alpha=0.8)
# ax.scatter()
scat = ax.scatter(
    x=asteroid_locations[:,0],
    y=asteroid_locations[:,1],
    # plot each point as a big circle???
    c="r", # color blue
    s=500,
    # circle=plt.Circle(location, 0.5 , color='r')
)

# scat = ax.scatter(
#     # x is first column of asteroid_locations
#     x=asteroid_locations[:,0],
#     y=asteroid_locations[:,1],
#     c="b",
#     s=5,
#     label=f'v0 = {v0} m/s'
# )

# scat = ax.scatter(
#     t[0],
#     z[0],
#     c="b",
#     s=5,
#     label=f'v0 = {v0} m/s'
# )

# line2 = ax.plot(
#     t[0],
#     z2[0],
#     label=f'v0 = {v02} m/s'
# )[0]

# ax.set(
#     xlim=[0, 3],
#     ylim=[-4, 10],
#     xlabel='Time [s]',
#     ylabel='Z [m]'
# )
# ax.legend()


def update(frame):
    # for each frame, update the data stored on each artist.
    # x = t[:frame]
    # y = z[:frame]
    x = asteroid_locations[:frame,0]
    y = asteroid_locations[:frame,1]
    # update the scatter plot:
    data = np.stack([x, y]).T
    scat.set_offsets(data)
    # update the line plot:
    # line2.set_xdata(t[:frame])
    # line2.set_ydata(z2[:frame])
    # return (scat, line2)
    return scat



ani = animation.FuncAnimation(
    fig=fig,
    func=update,
    frames=40,
    interval=1000,
    blit=True,
    repeat=False,
    repeat_delay=3000
)

# def animate(i):
#     redDot.set_data(i, np.sin(i))
#     return redDot,

# ani.FuncAnimation(
#     fig,
#     func=update,
#     # frames=np.arange(0.0, TWOPI, 0.1),
#     frames=40,
#     interval=1000,
#     blit=True,
#     repeat=False
# )

# mywriter = animation.FFMpegWriter()
# ani.save('mymovie.mp4',writer=mywriter)
# ani.save("test.gif")
ani.save(
    'myAnimation.gif',
    writer='imagemagick',
    fps=3
)


plt.show()
# ani.save("test.mp4", fps=30, dpi=300)
```
